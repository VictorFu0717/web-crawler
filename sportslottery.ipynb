```python
import requests
import json
import pandas as pd

table = {
    "比賽":[],
    "客隊":[],
    "主隊":[],
    "結果":[]
    }


h = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"}

page = 0
while True:
    page = page + 1
    url = "https://www.sportslottery.com.tw/api/services/app/GameResult/GetPagedSearchResult?s=447&c=0&l=0&p=90&o=1&g="+str(page)+"&z=-480&ct=zh-tw"
    response = requests.get(url, verify = False, headers=h)
    html = json.loads(response.text)
    data  = html["result"]["l"]
    realpage = html["result"]["c"]
    if page == realpage + 1:
        break

    for p in data:
        game = p["ln"]
        name1= p["atn"]
        name2= p["htn"]
        if p["ss"] is None:
            result = None
        elif p["ss"]["99"] == [-1,-1]:
            result = "無效投注"
        else:
            result1 = p["ss"]["99"][0]
            result2 = p["ss"]["99"][-1]
            result  = str(result1)+":"+str(result2)
        # print(game)  
        # print(name1)
        # print(name2)
        # print(result)
        table["比賽"].append(game)
        table["客隊"].append(name1)
        table["主隊"].append(name2)
        table["結果"].append(result)

df = pd.DataFrame(table)
df.to_csv("sportslotteryresult.csv", encoding="utf-8", index = False)
```

```python
import requests
import json
import pandas as pd

table = {
    "比賽":[],
    "客隊":[],
    "主隊":[],
    "客隊賠率":[],
    "主隊賠率":[],
    "和局賠率":[],
    "客隊理論勝率":[],
    "主隊理論勝率":[],
    "和局理論勝率":[],
    "客隊參考勝率":[],
    "主隊參考勝率":[],
    "和局參考勝率":[],
}

for i in range(441,448):
    url = "https://blob.sportslottery.com.tw/apidata/Pre/ListBySport/"+ str(i) + ".json?1615269375000"
    response = requests.get(url)
    html = json.loads(response.text)
    

    for p in html:
        game  = p["ln"][0]
        name1 = p["atn"][0]+p["atn"][1]
        name2 = p["htn"][0]+p["htn"][1]
        n = len(p["ms"][0]["cs"][0])
        
        
        if n==2:
            odds1=p["ms"][0]["cs"][0][0]["o"]
            odds2=p["ms"][0]["cs"][0][-1]["o"]
            odds3= None
            if odds1=="" or odds2=="":
                x = None
                y = None
                z = None
                a = None
                b = None
                c = None
            else:
                x = round(1/float(odds1),2)
                y = round(1/float(odds2),2)
                z = None
                a = round(x/(x+y),2)
                b = round(y/(x+y),2)
                c = None
            
        
        
        if n == 3 :
            odds1=p["ms"][0]["cs"][0][0]["o"]
            odds2=p["ms"][0]["cs"][0][-1]["o"]            
            odds3=p["ms"][0]["cs"][0][1]["o"]
            if odds1=="" or odds2=="" or odds3=="":
                x = None
                y = None
                z = None
                a = None
                b = None
                c = None
            else:
                x=round(1/float(odds1),2)
                y=round(1/float(odds2),2)
                z=round(1/float(odds3),2)
                a=round(x/(x+y+z),2)
                b=round(y/(x+y+z),2)
                c=round(z/(x+y+z),2)     
        
        
        
        table["比賽"].append(game)
        table["客隊"].append(name1)
        table["客隊賠率"].append(odds1)
        table["主隊"].append(name2)
        table["主隊賠率"].append(odds2)
        table["和局賠率"].append(odds3)          
        table["客隊理論勝率"].append(x)
        table["主隊理論勝率"].append(y)
        table["和局理論勝率"].append(z)
        table["客隊參考勝率"].append(a)
        table["主隊參考勝率"].append(b)
        table["和局參考勝率"].append(c)


df = pd.DataFrame(table)
df.to_csv('sportslottery.csv', encoding="utf-8", index = False)

from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication #傳送附件
content = MIMEMultipart()  #建立MIMEMultipart物件
content["subject"] = "sportslottery"  #郵件標題
content["from"] = "a0224658939@gmail.com"  #寄件者
content["to"] = "a0224658939@gmail.com" #收件者
content.attach(MIMEText("sports"))  #郵件內容

att1 = MIMEText(open('sportslottery.csv').read(), 'utf-8')
att1['Content-Type'] = 'application/octet-stream'
att1["Content-Disposition"] = 'attachment; filename="sportslottery.csv"' 
content.attach(att1)
import smtplib
with smtplib.SMTP(host="smtp.gmail.com", port="587") as smtp:  # 設定SMTP伺服器
    try:
        smtp.ehlo()  # 驗證SMTP伺服器
        smtp.starttls()  # 建立加密傳輸
        smtp.login("a0224658939@gmail.com", "ckpjtplcgjjiwtiy")  # 登入寄件者gmail
        smtp.send_message(content)  # 寄送郵件
        print("Complete!")
    except Exception as e:
        print("Error message: ", e)
```
